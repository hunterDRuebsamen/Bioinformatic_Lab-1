```python
def filter_dataframe(df, chromosome, start_loc, end_loc):
    """
    Filter the dataframe for the given chromosome and between start_loc and end_loc.
    
    Parameters:
    - df: pandas.DataFrame - The dataframe to filter.
    - chromosome: int - The chromosome number to filter by.
    - start_loc: int - The start location to filter by.
    - end_loc: int - The end location to filter by.
    
    Returns:
    - pandas.DataFrame - The filtered dataframe.
    """
    # Filter based on the conditions
    filtered_df = df[(df['chromosome'] == chromosome) & (df['s_loc'] >= start_loc) & (df['s_loc'] <= end_loc)]
    
    return filtered_df
```
